```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings("ignore")
```

```python
print("Consider the built-in dataset in seaborn library")
data= sns.load_dataset("penguins")
data.head()
```

# Histogram

```python
sns.displot(data, x="flipper_length_mm", kind="hist", binwidth=5)
plt.show()
```

```python
print("Note that the we can also include kde curve using kde=True as input.")
sns.displot(data, x="flipper_length_mm", kind="hist", binwidth=5, kde=True)
plt.show()
```

Note that the kind parameter support three kind of plots; "hist", "kde" and "ecdf".

```python
print("In other circumstances, it may make more sense to specify the number of bins using 'bins' parameter, rather than their size.")
```

Note that hue variable also works in distplot class.

```python
sns.displot(data, x="flipper_length_mm", binwidth=5, hue="sex", kind="hist")
plt.show()
```

To change histogram from bar type to step type, we can use element="step" as input. 

```python
sns.displot(data, x="flipper_length_mm", binwidth=5, hue="sex", element="step", kind="hist")
plt.show()
```

If we want to rearrange the histogram side by side horizontally, then we can use multiple="dodge" as input parameter.

```python
sns.displot(data, x="flipper_length_mm", binwidth=5, hue="sex", multiple="dodge", kind="hist")
plt.show()
```

If want to normalize the data, we can use 'stat' paramater as input. By default, however, the normalization is applied to the entire distribution, so this simply rescales the height of the bars. By setting common_norm=False, each subset will be normalized independently. Note that stat="density" normalized such that the area under the histogram (i.e., the sum of the bar areas) integrates to 1. Whereas in stat="probability" the heights of the bars represent the proportion of observations in each bin relative to the total number of observations.

```python
sns.displot(data, x="flipper_length_mm", binwidth=5, hue="sex", element="step",kind="hist", stat="density")
plt.show()
```

```python
sns.displot(data, x="flipper_length_mm", binwidth=5, hue="sex", element="step",kind="hist", stat="probability")
plt.show()
```

```python
print("In this, each subset will be normalized independently as we are using 'common_norm' parameter.")
sns.displot(data, x="flipper_length_mm", binwidth=5, hue="sex", element="step",kind="hist", stat="probability", common_norm=False)
plt.show()
```

In addition to or with hue, we can also use "col" or "row" parameters to split the histograms and visualize separately. Furthermore, "col_wrap" parameter is used to set max plot in single row or number of subplot columns.

```python
sns.displot(data, x="flipper_length_mm", binwidth=5, hue="sex", multiple="dodge", col="island", kind="hist",col_wrap=2, height=3)
plt.show()
```

# Kernel density estimation

Note that kernel density estimation plot smooths the observations with a Gaussian kernel, producing a continuous density estimate.

```python
sns.displot(kind="kde", data=data, x="flipper_length_mm", height=4)
plt.show()
```

The ability of the KDE to accurately represent the data depends on the choice of smoothing bandwidth. An over-smoothed estimate might erase meaningful features, but an under-smoothed estimate can obscure the true shape within random noise. The easiest way to check the robustness of the estimate is to adjust the default bandwidth.

```python
sns.displot(kind="kde", data=data, x="flipper_length_mm", height=4, bw_adjust=0.25)
plt.show()
```

Note that we can also hue variable in kde plot. It provides separate density estimate for each level of variable. Also, fill parameter can be use to fill the area with color and alpha parameter to control the opacity.

```python
sns.displot(kind="kde", data=data, x="flipper_length_mm", hue='sex', height=4, bw_adjust=0.25, fill=True)
plt.show()
```
